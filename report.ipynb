```python
from src.generators import DirectionalProblemReducedMazeProblemGenerator
from glob import glob

problem_count = 3
d = DirectionalProblemReducedMazeProblemGenerator(problem_count=problem_count)
d.display_images(problem_count)

datapoints = []
classical_result_lengths = [len(result.plan.actions) for result in d.solve_each()]
for i in range(10, 101, 10):
    d.solve_all(program_lines=i)
    total = 0
    for j, file in enumerate(glob("tmp/plan.*")):
        with open(file, "r") as f:
            generalised_len = len(f.read().split("\n")) - 1
            total += generalised_len / classical_result_lengths[j]
    avg = total / problem_count
    datapoints.append((i, avg))
            
print(datapoints)
```
